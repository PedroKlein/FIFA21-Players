## Contrução dos DataFrames

```python
import pandas as pd
import numpy as np
```

```python
players_df = pd.read_csv('./data/players.csv')
players_df
```

```python
tags_df = pd.read_csv('./data/tags.csv')
tags_df
```

```python
rating_df = pd.read_csv('./data/rating.csv')
rating_df
```

```python
rating_by_player_df = rating_df[['sofifa_id', 'rating']].groupby(by=['sofifa_id']).mean()
rating_by_player_df
```

```python
nrating_by_player_df = rating_df[['sofifa_id', 'user_id']].groupby(by=['sofifa_id']).count()
nrating_by_player_df
```

```python
nrating_by_player_df.loc[158023]
```

```python
rating_by_player_df.loc[158023]
```

```python
players_df = players_df.set_index('sofifa_id')
```

```python
players_df
```

```python
players_df['rating'] = [rating_by_player_df.loc[x]['rating'] if x in rating_by_player_df.index else 0 for x in players_df.index]
```

```python
players_df['count'] = [nrating_by_player_df.loc[x]['user_id'] if x in nrating_by_player_df.index else 0 for x in players_df.index]
```

```python
players_df
```

### Pelo usuário

```python
rating_by_user_df = rating_df[rating_df.user_id == 4]
rating_by_user_df
```

```python
rating_by_user_df = rating_by_user_df[['sofifa_id', 'rating']].copy().set_index('sofifa_id')
rating_by_user_df
```

```python
player_by_user_df = players_df.loc[rating_by_user_df.index][['name', 'rating', 'count']].copy()
player_by_user_df = player_by_user_df.rename(columns={'rating': 'global_rating'})
player_by_user_df
```

```python
player_by_user_df['rating'] = rating_by_user_df['rating']
```

```python
player_by_user_df.sort_values(by='rating', ascending=False).head(20)
```

### Melhores por posição


```python
players_per_pos_df = players_df[(players_df.player_positions.str.contains('ST')) & (players_df['count'] > 1000)]
players_per_pos_df.sort_values(by='rating', ascending=False).head(10)
```
